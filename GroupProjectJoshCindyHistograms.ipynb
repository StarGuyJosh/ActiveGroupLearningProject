```python
# START OF CODE FOR HISTOGRAM ANALYSIS

%matplotlib inline
import rasterio
import numpy
from matplotlib import pyplot as plt

# We load all of our imagery here
image_file_low1027 = "../Satellite Image Analysis Data/low_1027.tif"
my_image_low1027 = rasterio.open(image_file_low1027)
image_file_low1103 = "../Satellite Image Analysis Data/low_1103.tif"
my_image_low1103 = rasterio.open(image_file_low1103)
image_file_low1110 = "../Satellite Image Analysis Data/low_1110.tif"
my_image_low1110 = rasterio.open(image_file_low1110)
image_file_low1118 = "../Satellite Image Analysis Data/low_1118.tif"
my_image_low1118 = rasterio.open(image_file_low1118)
image_file_high1027 = "../Satellite Image Analysis Data/high_1027.tif"
my_image_high1027 = rasterio.open(image_file_high1027)
image_file_high1103 = "../Satellite Image Analysis Data/high_1103.tif"
my_image_high1103 = rasterio.open(image_file_high1103)
image_file_high1110 = "../Satellite Image Analysis Data/high_1110.tif"
my_image_high1110 = rasterio.open(image_file_high1110)
image_file_high1118 = "../Satellite Image Analysis Data/high_1118.tif"
my_image_high1118 = rasterio.open(image_file_high1118)

# We parse out the bands into individual variable names
# low bands
ndvi_l1 = my_image_low1027.read()
ndvi_l2 = my_image_low1103.read()
ndvi_l3 = my_image_low1110.read()
ndvi_l4 = my_image_low1118.read()

# high bands
ndvi_h1 = my_image_high1027.read()
ndvi_h2 = my_image_high1103.read()
ndvi_h3 = my_image_high1110.read()
ndvi_h4 = my_image_high1118.read()

# We create twoxtwo subplots and unpack the output array
fig, (ax1, ax2) = plt.subplots(2, 2, sharey=True)

# Next we put in our data
ndvi_data_l1 = ndvi_l1[numpy.not_equal(ndvi_l1, my_image_low1027.nodata)]
ndvi_data_l2 = ndvi_l2[numpy.not_equal(ndvi_l2, my_image_low1103.nodata)]
ndvi_data_l3 = ndvi_l3[numpy.not_equal(ndvi_l3, my_image_low1110.nodata)]
ndvi_data_l4 = ndvi_l4[numpy.not_equal(ndvi_l4, my_image_low1118.nodata)]
ndvi_data_h1 = ndvi_h1[numpy.not_equal(ndvi_h1, my_image_high1027.nodata)]
ndvi_data_h2 = ndvi_h2[numpy.not_equal(ndvi_h2, my_image_high1103.nodata)]
ndvi_data_h3 = ndvi_h3[numpy.not_equal(ndvi_h3, my_image_high1110.nodata)]
ndvi_data_h4 = ndvi_h4[numpy.not_equal(ndvi_h4, my_image_high1118.nodata)]

# Here we set up our NDVI histogram layers
bin_number = 50

ax1[0].hist(ndvi_data_h1,  color='darkblue', bins=bin_number, label='High', alpha=1)
ax1[1].hist(ndvi_data_h2,  color='darkred',bins=bin_number, label='High', alpha=1)
ax2[0].hist(ndvi_data_h3,  color='darkgreen',bins=bin_number, label='High', alpha=1)
ax2[1].hist(ndvi_data_h4,  color='gold',bins=bin_number, label='High', alpha=1)

ax1[0].hist(ndvi_data_l1, color='lightblue', bins=bin_number, label='Low', alpha=0.5)
ax1[1].hist(ndvi_data_l2, color='lightpink', bins=bin_number, label='Low', alpha=0.5)
ax2[0].hist(ndvi_data_l3,  color='lightgreen', bins=bin_number, label='Low', alpha=0.5)
ax2[1].hist(ndvi_data_l4, color='yellow', bins=bin_number, label='Low', alpha=0.5)


# Next we set up our plot titles 
ax1[0].set_title('10/27 NDVI Elevation Histogram')
ax1[1].set_title('11/03 NDVI Elevation Histogram')
ax2[0].set_title('11/10 NDVI Elevation Histogram')
ax2[1].set_title('11/18 NDVI Elevation Histogram')

# Here we make a title for the plot and add a legend
plt.suptitle('NDVI Histograms of High and Low Elevation Areas in the Blue Ridge Mountains', y=1.05)  
ax1[0].legend(loc='upper right')
ax1[1].legend(loc='upper right')
ax2[0].legend(loc='upper right')
ax2[1].legend(loc='upper right')

# Finally, we make sure to have space between the plots and then export it for analysis
fig.tight_layout()
fig.savefig("ndvi_histogram_time_series.png", dpi=200, bbox_inches='tight', pad_inches=0.7)
```
